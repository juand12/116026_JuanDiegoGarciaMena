```python
# ejercicio 11 taller
x=('AEIOUÁÉÍÓÚ')
y=('ÁÉÍÓÚáéíóú')
esp=(' ')
z=['and', 'in','elif','or','is','else','if','import','print','return','for','while','not']
a=input('digite cadena de texto: ')
conv=0
convt=0
cond=0
cones=0
conre=0
for i in a: # se itera a traves de la cadena de texto
    if i in x: # se mira si cada letra esta en una lista creada 
        conv=conv+1
print('la cantidad de vocales en mayuscula son: ',conv)
for s in a:   # se itera a traves de la cadena de texto
    if s in y:
        convt=convt+1
print('la cantidad de letras con tilde son: ',convt)
for c in a:
    if c.isdigit():  # se mira si tiene digitos es decir numeros
        cond=cond+1
print('la cantidad de digitos es: ',cond)
for e in a:
    if e==esp: # se creo una cadena de espacio
        cones=cones+1
print('la cantidad de espacios es: ',cones)
l=a.split() # se hace la cadena una lista para poder comparar con otra
for d in l:
    if d in z:
        conre=conre+1
print('la cantidad de palabras reservadas es: ',conre)
```

```python
#ejercicio 18 taller
m=eval(input('insertar matriz cuadrada: '))
print('tu matriz: ')
for i in m:  #se muestra la lista como matriz
    print(i)
a=0
j=[]
z=[]
print('la matriz triangular asociada es: ')
for s in m: # se cogen porciones de las sublistas, primero todo y va descendiendo
    z.append(s[a:len(m)+1])
    a=a+1
print(z)
for h in z:  # ya se tiene la matriz se cuentan los unos
    g=h.count(1)
    j.append(g)
k=max(j)
l=1
for x in j: # se observa cual fila tiene mas unos
    if x==k:
        print('la fila con mas UNOS es la fila número:' )
        print(l)
    l=l+1
print('de la matriz triangular asociada')
```
